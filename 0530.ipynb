```python
import nltk
from nltk.corpus import PlaintextCorpusReader
corpus_root= '/content/drive/MyDrive/0523txt'
wordlists=PlaintextCorpusReader(corpus_root,'.*')

books = wordlists.fileids()
print(books)
```

```python
import nltk
testText= wordlists.words('Emma watson UN.txt')
single= nltk.text.Text(testText)

findlist=["Who","a","female","we"]
single.dispersion_plot(findlist)
```

```python
from threading import Condition
books= wordlists.fileids()

cfd = nltk.ConditionalFreqDist((fileid,word)
   for fileid in books
   for word in wordlists.words(fileid))
findlist=["can","could","might","may","will"]
cfd.tabulate(Conditions=books,samples=findlist)
```

```python
cfd = nltk.ConditionalFreqDist(
    (target,fileid[:7])
     for fileid in wordlists.fileids()
     for w in wordlists.words(fileid)
     for target in["can","could","will"]
    if w.lower()==target)
cfd.plot()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# doc: https://amueller.github.io/word_cloud/generated/wordcloud.WordCloud.html

#讀取文本
file_text = open('/content/drive/MyDrive/0523txt/Taylor2022.txt','r').read()


# background_color 背景顏色  #可以用color hex https://htmlcolorcodes.com/
# width 寬
# height 高
# margin 圖片邊緣
# max_font_size 設定文字中最大的大小是多少
# min_font_size
# max_words  文字雲顯示最大詞數
# font_path="/tmp/DFFT_C7.ttc", #設置字體


word1 = WordCloud(
    background_color="black",
    width=2000,
    height=1000,
    margin= 10,
).generate(file_text)


plt.imshow(word1)
plt.axis("off")
plt.show()
```

```python
from wordcloud import STOPWORDS
print(STOPWORDS)
```

```python
f = open('/content/drive/MyDrive/0523txt/Taylor2022.txt','r').read()

stop = ["thing", "things", "one","s","us","t"] + list(STOPWORDS)
wordcloud = WordCloud(
    background_color="#22DFD3",
    width=3000,
    height=2000,
    margin=12,
    stopwords=stop
).generate(f)


plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
